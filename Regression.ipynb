**What is Regression**

Regression is one of the main types of supervised learning in machine learning. In regression, the training set contains observations (also called features)  and their associated continuous target values. The process of regression has 2 phases.


*   The first phase is explirubg the relationships between the observations and the targets. This is the training phase. 
*   The second phase is using the patterns from the first phase to generate the raget for a future observation. This is the prediction phase.

The main difference between regression and classification is that the output values in regression are continuous, while in classification they are discrete. This leads to different applications areas for these 2 supervised learning methods. Classification is bassically used to determine desired memberships or characteristics. Some of the projects are email being spam or not, newsgroup tocpis or ad click-through. On the other hand regression mainly involves estimating an outcome for forecasting a response.

Some of the machine learning regression problems are


*   Predicting house prices based on location, square footage, number of bedrooms and bathrooms
*   Estimating power consumption based on information about a system's processes and memory


*   forecasting demand in retail
*   Predicting stock prices.
